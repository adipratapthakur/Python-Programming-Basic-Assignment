```python
#1.	Write a Python Program to Find the Factorial of a Number?

def factorial():
    '''
    this function will calculate the factorial of a given number.
    '''
    try:
        num = int(input("Enter the number: "))
        
        if num<0:
            print("\nFactorial of negative number does not exists.")
        elif num==0 or num==1:
            print("\nFactorial of {} is 1.".format(num))
        else:
            result = 1
            for i in range(2,num+1):
                result *= i
            print("\nFactorial of {} is {}.".format(num,result))
    except Exception as e:
        print("\nSome exception occurred: ",e)
        
factorial()
```

```python
#2.	Write a Python Program to Display the multiplication Table?

def printTable():
    '''
    this function will print the multiplication table of a given number.
    '''
    try:
        num = int(input("Enter the number: "))
        
        for i in range(1,11):
            print("{} x {} = {}".format(num,i,num*i))
        
    except Exception as e:
        print("\nSome exception occurred: ",e)
        
printTable()
```

```python
#3.	Write a Python Program to Print the Fibonacci sequence?

def fibonacci():
    '''
    this function will print n terms of a fibonacci sequence.
    '''
    try:
        n_terms = int(input("How many number of terms to print of fibonacci sequence: "))
        num1 = 0
        num2 = 1
        
        for i in range(n_terms):
            print(num1)
            num1, num2 = num2 , num1+num2
        
    except Exception as e:
        print("\nSome exception occurred: ",e)
        
fibonacci()
```

```python
#4.	Write a Python Program to Check Armstrong Number?

def checkArmstrong():
    '''
    this function will check whether given number is an armstrong number or not.
    '''
    try:
        num = int(input("Enter the number: "))
        
        temp = str(num)
        length = len(temp)
        temp = int(temp)
        result = 0
        
        while temp>0:
            remainder = temp%10
            result += remainder**length
            temp //= 10
        
        if num==result:
            print("\n{} is an armstrong number.".format(num))
        else:
            print("\n{} is not an armstrong number.".format(num))
            
    except Exception as e:
        print("\nSome exception occurred: ",e)
        
checkArmstrong()
```

```python
#5.	Write a Python Program to Find Armstrong Number in an Interval?

def printArmstrong(num):
    '''
    this function will print all armstrong number in an interval.
    '''
    try:
        temp = str(num)
        length = len(temp)
        temp = int(temp)
        result = 0
        
        while temp>0:
            remainder = temp%10
            result += remainder**length
            temp //= 10
        
        if num==result:
            print("\n{} is an armstrong number.".format(num))
        else:
            print("\n{} is not an armstrong number.".format(num))
    except Exception as e:
        print("\nSome exception occurred: ",e)
        
try:
    lb = int(input("Enter the Lower bound of interval(inclusive): "))
    ub = int(input("Enter the Upper bound of interval(inclusive): "))
    
    for i in range(lb,ub+1):
        printArmstrong(i)
        
except Exception as e:
    print("\nSome exception occurred: ",e)
```

```python
#6.	Write a Python Program to Find the Sum of Natural Numbers?

def sumNatural():
    '''
    this function will give the sum of first n natural numbers.
    '''
    try:
        num = int(input("Enter the number to find sum of n natural numbers: "))
        
        result = int((num*(num+1))/2)
            
        print("\nSum of first {} natural numbers is: {}.".format(num,result))
        
    except Exception as e:
        print("\nSome exception occurred: ",e)
        
sumNatural()
```
